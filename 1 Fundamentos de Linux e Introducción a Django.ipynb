## Estructura de un proyecto en Django

### Archivo de Proyecto

- manage.py: es un archivo que contiene muchas funciones que ter permite interactuar con Django
- config/__init__.py: un archivo vacio que le informa a python que este directorio/carpeta deeria ser conswiderado como un paquete de python
- config/settings.py : configuracion o ajustes para el proyecto en django
-config/urls.py:las declaraciones de los urls para el proyecto django. Es como una tabla de contenidos con los endpoints del sitio web
-config.asgi.py: un punto de entrada para servidores web compatibles con ASGI para desplegar el proyecto

### Archivos de Aplicacion

- admin.py: registrar tus modelos relacionados con el django admin
- apps.py: contiene el codigo de la configuracion de la app
- models.py: define tus modelos aqui. Son las tablas en la base de datos para almacenar la informacion y definer sus relaciones
- test.py: escribir tests para la app
- view.py: definir la informacion que sera mandada a los templates del frontend.

import os

from django import get_version

from django.conf import settings

from django.shortcuts import render


def home(request):
    context = {
        "debug": settings.DEBUG,
        "django_ver": get_version() + " PROBANDO CAMBIOS",
        "python_ver": os.environ["PYTHON_VERSION"] + " MAS CAMBIOS",
    }

    return render(request, "pages/home.html", context)
